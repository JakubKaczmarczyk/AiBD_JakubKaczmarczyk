Zadanie 2 - wykresy 
W zadaniu drugim wykreśliłem 3 wykresy, co pozwoliło mi przypomnieć sobie jak korzystać z pakietu matplotlib.

```python
fun_kw = lambda x: x**2

wek1 = np.linspace(-1,1,1000)
wek2 = np.linspace(-6,6,1000)
wek3 = np.linspace(0,5,1000)

def plotuj(x,fx):
    plt.plot(x,fx)
    plt.title("wykres funkcji x^2+5")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.legend("funkcja f(x)")
    plt.show()

    
plotuj(wek1, fun_kw(wek1))
plotuj(wek2, fun_kw(wek2))
plotuj(wek3, fun_kw(wek3))

```
