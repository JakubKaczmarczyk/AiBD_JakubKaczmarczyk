```python
import matplotlib.pyplot as plt
import numpy as np

def fun_kw(x):
    return x**2+5

wek1 = np.arange(-1,1,0.01)
wek2 = np.arange(-6,6,0.01)
wek3 = np.arange(0,5,0.01)

def plotuj(x,fx):
    plt.plot(x,fx)
    plt.title("wykres funkcji x**2+5")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.legend("funkcja f(x)")
    plt.show()

    
plotuj(wek1, fun_kw(wek1))
plotuj(wek2, fun_kw(wek2))
plotuj(wek3, fun_kw(wek3))

# Jakub Kaczmarczyk, 400397
```
