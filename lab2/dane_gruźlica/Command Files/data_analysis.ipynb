Przetworzone dane poddałem analizie, obliczając sumy zakażeń w każdym z krajów na przedziale badanych lat 1989-2008. Wyniki wyświetliłem na wykresach.

```python
# Plots
import matplotlib.pyplot as plt
import csv
import os
import pandas as pd
import numpy as np

# Access to file
tb_path = os.getcwd()
tb_path = tb_path[:-14]+"\Analysis Data\\tbv2.csv"

df = pd.read_csv(tb_path)

# creating statistics 
countries_set = set(df['Country'])
countries_list = list(countries_set)
countries_list.sort()
sum_cases = np.zeros(len(countries_list))
index_range = list(np.arange(0,3691)) + list(np.arange(3714, len(df['Country'])))
for i in range(0,len(countries_list)):
    for j in index_range:
        if df['Country'][j] == countries_list[i]:
            sum_cases[i] += df['New cases'][j]

            
plt.figure(figsize=(20,10))
plt.bar(np.arange(50),sum_cases[:50])
plt.xticks(np.arange(50),countries_list[:50])
plt.title('Liczba zakażeń gruźlicą na przestrzeni lat 1989-2008')
plt.xlabel('Kraj')
plt.ylabel('Liczba zakażeń')
plt.ylim([0,5e6])
plt.savefig('Countries_plot_p1')
# plt.show()
plt.figure(figsize=(20,10))
plt.bar(np.arange(50),sum_cases[50:100])
plt.xticks(np.arange(50),countries_list[50:100])
plt.title('Liczba zakażeń gruźlicą na przestrzeni lat 1989-2008')
plt.xlabel('Kraj')
plt.ylabel('Liczba zakażeń')
plt.ylim([0,5e6])
plt.savefig('Countries_plot_p2')
# plt.show()
plt.figure(figsize=(20,10))
plt.bar(np.arange(50),sum_cases[100:150])
plt.xticks(np.arange(50),countries_list[100:150])
plt.title('Liczba zakażeń gruźlicą na przestrzeni lat 1989-2008')
plt.xlabel('Kraj')
plt.ylabel('Liczba zakażeń')
plt.ylim([0,5e6])
plt.savefig('Countries_plot_p3')
# plt.show()
plt.figure(figsize=(20,10))
plt.bar(np.arange(62),sum_cases[150:212])
plt.xticks(np.arange(62),countries_list[150:212])
plt.title('Liczba zakażeń na gruźlicą przestrzeni lat 1989-2008')
plt.xlabel('Kraj')
plt.ylabel('Liczba zakażeń')
plt.ylim([0,5e6])
plt.savefig('Countries_plot_p4')
# plt.show()
```
